## Data Description

- Vol : The total sum of count collected within a 15 minute increments.
- Yr : The two digit year portion of the date when the count was conducted.
- M : The two digit month portion of the date when the count was conducted.
- D : The two digit day portion of the date when the count was conducted.
- HH : The two digit hour portion of the time when the count was conducted.
- MM : The two digit start minute portion of the time when the count was conducted.
- street : The 'On Street' where the count took place.
- Boro : Lists which of the five administrative divisions of New York City the location is within, written as a word.
- RequestID : An unique ID that is generated for each counts request.
- SegmentID : The ID that identifies each segment of a street in the LION street network version 14.
- WktGeom : A text markup language for representing vector geometry objects on a map and spatial reference systems of spatial objects.
- fromSt : The 'From Street' where the count took place.
- toSt : The 'To Street' where the count took place.
- Direction : The text-based direction of traffic where the count took place.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter
```

```python
# 1. Hourly Traffic Patterns
hourly_traffic = df.groupby('HH')['Vol'].sum().reset_index()

plt.figure(figsize=(14, 7))
sns.barplot(x='HH', y='Vol', data=hourly_traffic, color='steelblue')
plt.title('Total Traffic Volume by Hour of Day', fontsize=16)
plt.xlabel('Hour of Day', fontsize=14)
plt.ylabel('Total Volume', fontsize=14)
plt.xticks(range(0, 24))
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Format y-axis to show numbers in millions
def millions(x, pos):
    return f'{x/1e6:.1f}M'
    
plt.gca().yaxis.set_major_formatter(FuncFormatter(millions))
plt.tight_layout()
plt.show()
```

```python


# 1. Monthly Traffic Patterns
# First, extract month from datetime_hour
df['month'] = df['datetime_hour'].dt.month
df['month_name'] = df['datetime_hour'].dt.month_name()

monthly_traffic = df.groupby(['month', 'month_name'])['Vol'].sum().reset_index()
monthly_traffic = monthly_traffic.sort_values('month')

plt.figure(figsize=(14, 7))
sns.barplot(x='month_name', y='Vol', data=monthly_traffic, palette='coolwarm')
plt.title('Total Traffic Volume by Month', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Total Volume', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.gca().yaxis.set_major_formatter(FuncFormatter(millions))
plt.tight_layout()
plt.show()

# 2. Peak Traffic Hours by Borough
# Extract hour for analysis
df['hour'] = df['datetime_hour'].dt.hour

hourly_borough = df.groupby(['Boro', 'hour'])['Vol'].sum().reset_index()

plt.figure(figsize=(15, 10))
sns.lineplot(x='hour', y='Vol', hue='Boro', data=hourly_borough, palette='tab10', linewidth=2.5)
plt.title('Hourly Traffic Volume by Borough', fontsize=18)
plt.xlabel('Hour of Day', fontsize=14)
plt.ylabel('Total Volume', fontsize=14)
plt.xticks(range(0, 24))
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Borough', fontsize=12, title_fontsize=14)
plt.tight_layout()
plt.show()

# Find peak hour for each borough
peak_hours = hourly_borough.loc[hourly_borough.groupby('Boro')['Vol'].idxmax()]
peak_hours = peak_hours.sort_values('Vol', ascending=False)
print("Peak Traffic Hours by Borough:")
print(peak_hours[['Boro', 'hour', 'Vol']])

# Find overall peak hour
overall_peak = hourly_traffic.loc[hourly_traffic['Vol'].idxmax()]
print(f"\nOverall Peak Traffic Hour: {overall_peak['HH']} with {overall_peak['Vol']:,} vehicles")

# 3. Time Series Analysis - Daily Trend
df['date'] = df['datetime_hour'].dt.date
daily_volume = df.groupby('date')['Vol'].sum().reset_index()
daily_volume['date'] = pd.to_datetime(daily_volume['date'])
daily_volume = daily_volume.sort_values('date')

plt.figure(figsize=(16, 8))
sns.lineplot(x='date', y='Vol', data=daily_volume, linewidth=2)
plt.title('Daily Traffic Volume Over Time', fontsize=18)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Total Volume', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, p: f'{x/1e6:.1f}M'))
plt.tight_layout()
plt.show()

# 4. Weekly Traffic Patterns
df['week'] = df['datetime_hour'].dt.isocalendar().week
weekly_volume = df.groupby('week')['Vol'].sum().reset_index()

plt.figure(figsize=(16, 8))
sns.lineplot(x='week', y='Vol', data=weekly_volume, marker='o', linewidth=2)
plt.title('Weekly Traffic Volume Over Time', fontsize=18)
plt.xlabel('Week Number', fontsize=14)
plt.ylabel('Total Volume', fontsize=14)
# Fix: Remove duplicate ylabel and restore grid
plt.grid(True, linestyle='--', alpha=0.7)
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, p: f'{x/1e6:.1f}M'))
plt.tight_layout()
plt.show()
# 5. Heatmap: Hour of Day vs Day of Week
# Fix: Create day of week features first
df['day_of_week'] = df['datetime_hour'].dt.day_name()
df['day_of_week_num'] = df['datetime_hour'].dt.dayofweek

# First verify we have data
print(f"Number of day/hour combinations: {df.groupby(['day_of_week_num', 'hour']).ngroups}")

# Create the heatmap data
hourly_daily = df.groupby(['day_of_week_num', 'hour'])['Vol'].sum().reset_index()
hourly_daily_pivot = hourly_daily.pivot(index='day_of_week_num', columns='hour', values='Vol')

# Check if pivot table has data
print(f"Pivot table shape: {hourly_daily_pivot.shape}")
print("Sample of pivot table data:")
print(hourly_daily_pivot.iloc[:3, :5])  # Show first 3 rows, 5 columns

plt.figure(figsize=(16, 8))
sns.heatmap(hourly_daily_pivot, cmap='YlOrRd', robust=True, 
            yticklabels=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Traffic Volume Heatmap: Day of Week vs Hour of Day', fontsize=18)
plt.xlabel('Hour of Day', fontsize=14)
plt.ylabel('Day of Week', fontsize=14)
plt.tight_layout()
plt.show()
```
